### O que vamos procurar nos gráficos?
Morfologia IR vs. RED: Ver se o Infravermelho está realmente entregando a "assinatura" da pressão sanguínea melhor que o Vermelho.

Estabilidade Temporal (Jitter): Verificar se o intervalo entre as amostras é constante (essencial para o Alpha 1).

Relação de Fase: Entender como a onda de luz se comporta antes de tentarmos "transformá-la" em sinal elétrico de ECG.

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import json
```

```python
# Carregar o seu arquivo
df = pd.read_csv('teste1_dados_400hz_v8.csv')
```

```python
# Parse das listas JSON
df['ir_waveform'] = df['ir_waveform'].apply(json.loads)
df['red_waveform'] = df['red_waveform'].apply(json.loads)
```

```python
print(df.columns)
print(df.head())
```

```python
# Concatena tudo
ir_data = np.concatenate(df['ir_waveform'].values).astype(float)
red_data = np.concatenate(df['red_waveform'].values).astype(float)
```

```python
# Normaliza Z-score
ir_norm = (ir_data - ir_data.mean()) / ir_data.std()
red_norm = (red_data - red_data.mean()) / red_data.std()
```

```python
# Plota
fig, axes = plt.subplots(2, 1, figsize=(15, 8))

axes[0].plot(ir_norm[:2000], color='darkred', linewidth=0.8)
axes[0].set_title('IR Normalizado')
axes[0].set_xlabel('Amostras')
axes[0].set_ylabel('Z-score')
axes[0].grid(True, alpha=0.3)

axes[1].plot(red_norm[:2000], color='red', linewidth=0.8)
axes[1].set_title('RED Normalizado')
axes[1].set_xlabel('Amostras')
axes[1].set_ylabel('Z-score')
axes[1].grid(True, alpha=0.3)

plt.show()
```
